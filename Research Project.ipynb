```python
import matplotlib.pyplot as plt
import concurrent.futures
import time

def process_file(filename):
    # Read file contents
    with open(filename, 'r') as f:
        content = f.read()

    # Process file contents (e.g., convert to uppercase)
    processed_content = content.upper()

    # Write processed content to a new file
    with open('output.txt', 'w') as f:
        f.write(processed_content)

def run_asynchronous_task(filename):
    start_time = time.time()
    process_file(filename)
    return time.time() - start_time

def run_parallel_task(filename):
    start_time = time.time()
    with concurrent.futures.ProcessPoolExecutor() as executor:
        executor.submit(process_file, filename)
    return time.time() - start_time

def main():
    # Prepare data
    filenames = ['input.txt'] * 3  # Assuming input.txt is your input file
    asynchronous_times = []
    parallel_times = []

    # Run tasks and measure execution time
    for filename in filenames:
        asynchronous_time = run_asynchronous_task(filename)
        parallel_time = run_parallel_task(filename)
        asynchronous_times.append(asynchronous_time)
        parallel_times.append(parallel_time)
        print(f"Filename: {filename}")
        print(f"Asynchronous Execution Time: {asynchronous_time} seconds")
        print(f"Parallel Execution Time: {parallel_time} seconds")

    # Plotting the graph
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(filenames)), asynchronous_times, label='Asynchronous')
    plt.plot(range(len(filenames)), parallel_times, label='Parallel')
    plt.xticks(range(len(filenames)), [f"File {i+1}" for i in range(len(filenames))])
    plt.title('Asynchronous vs Parallel Execution Time')
    plt.xlabel('File')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import multiprocessing
import time
import math

# Define a CPU-bound operation
def cpu_bound_operation(n):
    # Perform some CPU-intensive computation (e.g., calculate factorial)
    result = math.factorial(n)
    return result

# Function to run CPU-bound operation asynchronously
def run_asynchronous_task(n):
    start_time = time.time()
    with multiprocessing.Pool() as pool:
        result = pool.apply(cpu_bound_operation, (n,))
    return time.time() - start_time

# Function to run CPU-bound operation in parallel
def run_parallel_task(n):
    start_time = time.time()
    with multiprocessing.Pool() as pool:
        result = pool.apply(cpu_bound_operation, (n,))
    return time.time() - start_time

# Main function to compare execution times and plot the graph
def main():
    # Varying input sizes
    input_sizes = [5]

    # Lists to store execution times
    asynchronous_times = []
    parallel_times = []

    # Measure execution times for each input size
    for n in input_sizes:
        asynchronous_time = run_asynchronous_task(n)
        parallel_time = run_parallel_task(n)
        asynchronous_times.append(asynchronous_time)
        parallel_times.append(parallel_time)
        print(f"Input size: {n}")
        print(f"Asynchronous Execution Time: {asynchronous_time} seconds")
        print(f"Parallel Execution Time: {parallel_time} seconds")

    # Plotting the graph
    plt.figure(figsize=(10, 6))
    plt.plot(input_sizes, asynchronous_times, label='Asynchronous')
    plt.plot(input_sizes, parallel_times, label='Parallel')
    plt.title('Asynchronous vs Parallel Execution Time (CPU Bound Operation)')
    plt.xlabel('Input Size')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import aiohttp
import asyncio
import concurrent.futures
import requests
import time
import matplotlib.pyplot as plt

# Define a list of URLs to fetch data from
URLS = [
    'https://place.dog/300/200',
    'https://place.dog/400/300',
    'https://place.dog/500/400',
    'https://place.dog/600/500',
    'https://place.dog/700/600'
]

# Asynchronous concurrency to fetch data from URLs
async def asynchronous_fetch(url):
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as response:
            return await response.read()

async def asynchronous_concurrency():
    start_time = time.time()
    await asyncio.gather(*[asynchronous_fetch(url) for url in URLS])
    return time.time() - start_time

# Parallel concurrency to fetch data from URLs
def parallel_fetch(url):
    response = requests.get(url)
    return response.content

def parallel_concurrency():
    start_time = time.time()
    with concurrent.futures.ThreadPoolExecutor() as executor:
        list(executor.map(parallel_fetch, URLS))
    return time.time() - start_time

# Main function to compare execution times and plot the graph
async def main():
    async_times = []
    parallel_times = []

    for _ in range(5):  # Perform the comparison multiple times for more accurate results
        async_time = await asynchronous_concurrency()
        parallel_time = parallel_concurrency()

        async_times.append(async_time)
        parallel_times.append(parallel_time)

    # Plotting the graph
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, 6), async_times, label='Asynchronous')
    plt.plot(range(1, 6), parallel_times, label='Parallel')
    plt.title('Asynchronous vs Parallel Concurrency Performance')
    plt.xlabel('Iteration')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the main coroutine within the event loop
await main()
```

```python
import aiohttp
import asyncio
import concurrent.futures
import requests
import time
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO

# Define a list of URLs to fetch images from (Place Dog API)
URLS = [
    'https://place.dog/300/200',
    'https://place.dog/400/300',
    'https://place.dog/500/400',
    'https://place.dog/600/500',
    'https://place.dog/700/600'
]

# Asynchronous concurrency to fetch images from URLs
async def asynchronous_fetch(url):
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as response:
            data = await response.read()
            return Image.open(BytesIO(data))

async def asynchronous_concurrency():
    start_time = time.time()
    await asyncio.gather(*[asynchronous_fetch(url) for url in URLS])
    return time.time() - start_time

# Parallel concurrency to fetch images from URLs
def parallel_fetch(url):
    response = requests.get(url)
    data = response.content
    return Image.open(BytesIO(data))

def parallel_concurrency():
    start_time = time.time()
    with concurrent.futures.ThreadPoolExecutor() as executor:
        list(executor.map(parallel_fetch, URLS))
    return time.time() - start_time

# Main function to compare execution times and plot the graph
async def main():
    async_times = []
    parallel_times = []

    for _ in range(5):  # Perform the comparison multiple times for more accurate results
        async_time = await asynchronous_concurrency()
        parallel_time = parallel_concurrency()

        async_times.append(async_time)
        parallel_times.append(parallel_time)

    # Plotting the graph
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, 6), async_times, label='Asynchronous')
    plt.plot(range(1, 6), parallel_times, label='Parallel')
    plt.title('Asynchronous vs Parallel Concurrency Performance (I/O Bound)')
    plt.xlabel('Iteration')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the main coroutine within the event loop
await main()
```

```python
import asyncio
import concurrent.futures
import time
import matplotlib.pyplot as plt

# Define the data processing task (just a simple multiplication operation)
def data_processing_task(num):
    return num * num

# Asynchronous data processing function
async def asynchronous_data_processing(data):
    start_time = time.time()
    await asyncio.gather(*[asyncio.to_thread(data_processing_task, num) for num in data])
    return time.time() - start_time

# Parallel data processing function
def parallel_data_processing(data):
    start_time = time.time()
    with concurrent.futures.ThreadPoolExecutor() as executor:
        list(executor.map(data_processing_task, data))
    return time.time() - start_time

# Main function to compare execution times and plot the graph
async def main():
    dataset_sizes = [100, 200, 300, 400, 500]  # Different sizes of datasets
    asynchronous_times = []
    parallel_times = []

    for size in dataset_sizes:
        data = [i for i in range(size)]  # Generate sample data
        asynchronous_time = await asynchronous_data_processing(data)
        parallel_time = parallel_data_processing(data)
        asynchronous_times.append(asynchronous_time)
        parallel_times.append(parallel_time)

    # Plotting the graph
    plt.plot(dataset_sizes, asynchronous_times, label='Asynchronous')
    plt.plot(dataset_sizes, parallel_times, label='Parallel')
    plt.title('Asynchronous vs Parallel Data Processing Performance')
    plt.xlabel('Dataset Size')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the main coroutine within the event loop
await main()
```
